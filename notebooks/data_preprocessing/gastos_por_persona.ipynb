```python
# Diccionario de entidades 
mapa_entidades = {
    1: "Aguascalientes",
    2: "Baja California",
    3: "Baja California Sur",
    4: "Campeche",
    5: "Coahuila de Zaragoza",
    6: "Colima",
    7: "Chiapas",
    8: "Chihuahua",
    9: "Ciudad de México",
    10: "Durango",
    11: "Guanajuato",
    12: "Guerrero",
    13: "Hidalgo",
    14: "Jalisco",
    15: "Estado de México", 
    16: "Michoacán de Ocampo",
    17: "Morelos",
    18: "Nayarit",
    19: "Nuevo León",
    20: "Oaxaca",
    21: "Puebla",
    22: "Querétaro",
    23: "Quintana Roo",
    24: "San Luis Potosí",
    25: "Sinaloa",
    26: "Sonora",
    27: "Tabasco",
    28: "Tamaulipas",
    29: "Tlaxcala",
    30: "Veracruz de Ignacio de la Llave",
    31: "Yucatán",
    32: "Zacatecas"
}
```

```python
# Limpiamos la columna de gasto trimestral
df['gasto_tri'] = pd.to_numeric(df['gasto_tri'].astype(str).str.replace(',', '', regex=False).str.strip(), errors='coerce')

df['gasto_tri_factor'] = df['gasto_tri'] * df['factor']
# Calculamos el gasto promedio por entidad
gasto_por_entidad = df.groupby('entidad')['gasto_tri_factor'].mean()

# Reseteamos el indice
df_gasto_tri_por_entidad = gasto_por_entidad.reset_index()

# Modificamos nombre de columnas 
df_gasto_tri_por_entidad['entidad'] = df_gasto_tri_por_entidad['entidad'].map(mapa_entidades)

# Imprimimos resultado 
df_gasto_tri_por_entidad
```

```python
df_gasto_tri_por_entidad.rename(columns={'entidad': 'NOM_ENT', 'gasto_tri_factor': 'GASTO_PROM_TRI'}, inplace=True)
```

```python
df_gasto_tri_por_entidad.to_csv('gasto_promedio_trim_por_entidad_2024.csv', index = False)
```
