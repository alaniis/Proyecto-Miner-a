```python
import pandas as pd 
import os 
import glob
```

```python
ruta_archivo = '..\\data\\pobreza_2022_bueno\\pobreza.xlsx'
df = pd.read_excel(ruta_archivo, sheet_name=0, header = [0, 1])
```

```python
df.columns = ['_'.join(col).strip() for col in df.columns.values]
df
```

```python
# Lista de columnas que deseas eliminar
columnas_interes = ['Entidad federativa_Unnamed: 1_level_1','Municipio_Unnamed: 3_level_1',
                    'Pobreza_Carencias promedio\n2020', 'Vulnerables por carencia social_Carencias promedio\n2020',
                    'Vulnerables por ingreso_Porcentaje\n2020', 'Carencia por acceso a los servicios de salud_Carencias promedio\n2020',
                    'Carencia por acceso a la seguridad social_Carencias promedio\n2020']
# Crear un nuevo DataFrame con solo las columnas seleccionadas
df = df[columnas_interes].copy()
```

```python
df.rename(columns={
    'Entidad federativa_Unnamed: 1_level_1': 'NOM_ENT',
    'Municipio_Unnamed: 3_level_1': 'NOM_MUN',
    'Pobreza_Carencias promedio\n2020': 'POBR_CAR_PROM',
    'Vulnerables por carencia social_Carencias promedio\n2020': 'VUL_CAR_PROM',
    'Vulnerables por ingreso_Porcentaje\n2020': 'VUL_ING_POR',
    'Carencia por acceso a los servicios de salud_Carencias promedio\n2020': 'CAR_SALUD_PROM',
    'Carencia por acceso a la seguridad social_Carencias promedio\n2020': 'CAR_SEG_SOC_PROM'
}, inplace=True)
```
