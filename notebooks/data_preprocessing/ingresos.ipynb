```python
# Importamos las librerías necesarias
import pandas as pd 
```

```python
# Importamos el conjunto de datos
df = pd.read_csv('../../raw_data/trimestral_entity_income/conjunto_de_datos_ingresos_enigh2024_ns.csv')
df.head()
```

```python
# Mapeo de entidades federativas
mapa_entidades = {
    1: "Aguascalientes",
    2: "Baja California",
    3: "Baja California Sur",
    4: "Campeche",
    5: "Coahuila",
    6: "Colima",
    7: "Chiapas",
    8: "Chihuahua",
    9: "Ciudad de México",
    10: "Durango",
    11: "Guanajuato",
    12: "Guerrero",
    13: "Hidalgo",
    14: "Jalisco",
    15: "Estado de México", # 'México' en la lista, pero 'Estado de México' es menos ambiguo
    16: "Michoacán",
    17: "Morelos",
    18: "Nayarit",
    19: "Nuevo León",
    20: "Oaxaca",
    21: "Puebla",
    22: "Querétaro",
    23: "Quintana Roo",
    24: "San Luis Potosí",
    25: "Sinaloa",
    26: "Sonora",
    27: "Tabasco",
    28: "Tamaulipas",
    29: "Tlaxcala",
    30: "Veracruz",
    31: "Yucatán",
    32: "Zacatecas"
}
```

```python
df['ing_tri_factor'] = df['ing_tri'] * df['factor']
# Agrupamos por entidad numerica 
ing_tri_por_entidad = df.groupby('entidad')['ing_tri_factor'].mean()

df_ing_tri_por_entidad = ing_tri_por_entidad.reset_index() 

# Realizamos el mapeo de entidades 
df_ing_tri_por_entidad['entidad'] = df_ing_tri_por_entidad['entidad'].map(mapa_entidades)

df_ing_tri_por_entidad
```

```python
# Guardamos el DataFrame resultante en un archivo CSV
df_ing_tri_por_entidad.to_csv('../../data/processed_data/ing_tri_por_entidad_2024.csv', index = False)
```
