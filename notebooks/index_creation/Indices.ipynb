## Funciones de valor

```python
def minmax_norm(x):
    """
    Normaliza x en [0, 1] con min max.
    Devuelve (x_norm, xmin, xmax).
    """
    x = np.asarray(x, dtype=float)
    xmin = np.nanmin(x)
    xmax = np.nanmax(x)
    if xmax == xmin:
        return np.zeros_like(x), xmin, xmax
    return (x - xmin) / (xmax - xmin), xmin, xmax
```

```python
def convex_increasing(x, gamma=1.0, xmin=None, xmax=None):
    """
    Funcion convexa creciente del manual.
    vc = (exp(gamma * x_norm) - 1) / (exp(gamma) - 1)
    """
    x = np.asarray(x, dtype=float)
    if xmin is None or xmax is None:
        x_norm, xmin, xmax = minmax_norm(x)
    else:
        if xmax == xmin:
            x_norm = np.zeros_like(x)
        else:
            x_norm = (x - xmin) / (xmax - xmin)
    num = np.exp(gamma * x_norm) - 1.0
    den = np.exp(gamma) - 1.0
    return num / den
```

```python
def convex_decreasing(x, gamma=1.0, xmin=None, xmax=None):
    """
    Funcion convexa decreciente del manual.
    vd = (exp(gamma * (1 - x_norm)) - 1) / (exp(gamma) - 1)
    """
    x = np.asarray(x, dtype=float)
    if xmin is None or xmax is None:
        x_norm, xmin, xmax = minmax_norm(x)
    else:
        if xmax == xmin:
            x_norm = np.zeros_like(x)
        else:
            x_norm = (x - xmin) / (xmax - xmin)
    num = np.exp(gamma * (1.0 - x_norm)) - 1.0
    den = np.exp(gamma) - 1.0
    return num / den
```

```python
def concave_increasing(x, gamma=1.0, xmin=None, xmax=None):
    """
    Funcion concava creciente del manual.
    cc = 1 − vd
    """
    return 1.0 - convex_decreasing(x, gamma=gamma, xmin=xmin, xmax=xmax)
```

```python
def concave_increasing(x, gamma=1.0, xmin=None, xmax=None):
    """
    Funcion concava creciente del manual.
    cc = 1 − vd
    """
    return 1.0 - convex_decreasing(x, gamma=gamma, xmin=xmin, xmax=xmax)
```

```python
def concave_decreasing(x, gamma=1.0, xmin=None, xmax=None):
    """
    Funcion concava decreciente del manual.
    cd = 1 − vc
    """
    return 1.0 - convex_increasing(x, gamma=gamma, xmin=xmin, xmax=xmax)
```

```python
def bell(x, alpha=10.0, xc=None, xmin=None, xmax=None):
    """
    Funcion campana del manual.
    g = exp(-beta**2)
    beta = 100 * (x_norm - x_c_norm) / alpha
    """
    x = np.asarray(x, dtype=float)

    if xmin is None or xmax is None:
        x_norm, xmin_auto, xmax_auto = minmax_norm(x)
        xmin = xmin if xmin is not None else xmin_auto
        xmax = xmax if xmax is not None else xmax_auto
    else:
        if xmax == xmin:
            x_norm = np.zeros_like(x)
        else:
            x_norm = (x - xmin) / (xmax - xmin)

    if xc is None:
        xc = (xmin + xmax) / 2.0

    if xmax != xmin:
        x_c_norm = (xc - xmin) / (xmax - xmin)
    else:
        x_c_norm = 0.5

    beta = 100.0 * (x_norm - x_c_norm) / alpha
    return np.exp(-beta ** 2)
```

```python
def inverted_bell(x, alpha=10.0, xc=None, xmin=None, xmax=None):
    """
    Funcion campana invertida gi = 1 − g.
    """
    return 1.0 - bell(x, alpha=alpha, xc=xc, xmin=xmin, xmax=xmax)
```

```python
def logistic(x, alpha=10.0, xc=None, xmin=None, xmax=None):
    """
    Funcion logistica del manual.
    l = 1 / (1 + exp(-beta))
    beta = 100 * (x_norm - x_c_norm) / alpha
    """
    x = np.asarray(x, dtype=float)

    if xmin is None or xmax is None:
        x_norm, xmin_auto, xmax_auto = minmax_norm(x)
        xmin = xmin if xmin is not None else xmin_auto
        xmax = xmax if xmax is not None else xmax_auto
    else:
        if xmax == xmin:
            x_norm = np.zeros_like(x)
        else:
            x_norm = (x - xmin) / (xmax - xmin)

    if xc is None:
        xc = (xmin + xmax) / 2.0

    if xmax != xmin:
        x_c_norm = (xc - xmin) / (xmax - xmin)
    else:
        x_c_norm = 0.5

    beta = 100.0 * (x_norm - x_c_norm) / alpha
    return 1.0 / (1.0 + np.exp(-beta))
```

```python

def inverted_logistic(x, alpha=10.0, xc=None, xmin=None, xmax=None):
    """
    Funcion logistica invertida li = 1 − l.
    """
    return 1.0 - logistic(x, alpha=alpha, xc=xc, xmin=xmin, xmax=xmax)
```

```python
def aplicar_funcion_valor(serie, config):
    tipo = config.get("tipo")
    gamma = config.get("gamma", 1.0)
    alpha = config.get("alpha", 10.0)
    xc = config.get("xc")
    xmin = config.get("xmin")
    xmax = config.get("xmax")

    if tipo == "convexa_creciente":
        return convex_increasing(serie.values, gamma=gamma, xmin=xmin, xmax=xmax)
    if tipo == "concava_creciente":
        return concave_increasing(serie.values, gamma=gamma, xmin=xmin, xmax=xmax)
    if tipo == "convexa_decreciente":
        return convex_decreasing(serie.values, gamma=gamma, xmin=xmin, xmax=xmax)
    if tipo == "concava_decreciente":
        return concave_decreasing(serie.values, gamma=gamma, xmin=xmin, xmax=xmax)
    if tipo == "campana":
        return bell(serie.values, alpha=alpha, xc=xc, xmin=xmin, xmax=xmax)
    if tipo == "campana_invertida":
        return inverted_bell(serie.values, alpha=alpha, xc=xc, xmin=xmin, xmax=xmax)
    if tipo == "logistica":
        return logistic(serie.values, alpha=alpha, xc=xc, xmin=xmin, xmax=xmax)
    if tipo == "logistica_invertida":
        return inverted_logistic(serie.values, alpha=alpha, xc=xc, xmin=xmin, xmax=xmax)

    raise ValueError(f"Tipo de funcion de valor no reconocido: {tipo}")
```

```python
import numpy as np

def weber_fechner_bins(serie, n_clases=5, epsilon=1e-6):
    """
    Genera n_clases intervalos tipo Weber Fechner en el rango observado.
    Trabaja con logaritmos sobre la serie desplazada para evitar log(0).
    Devuelve un array de longitud n_clases + 1 con los puntos de corte.
    """
    s = pd.Series(serie).dropna()
    if s.empty:
        raise ValueError("La serie no tiene datos no nulos")

    x_min = s.min()
    x_max = s.max()

    if x_max == x_min:
        return np.linspace(x_min, x_max + epsilon, n_clases + 1)

    shift = 0.0
    if x_min <= 0:
        shift = abs(x_min) + epsilon

    I_min = x_min + shift + epsilon
    I_max = x_max + shift

    S_min = np.log(I_min)
    S_max = np.log(I_max)

    S_edges = np.linspace(S_min, S_max, n_clases + 1)

    I_edges = np.exp(S_edges) - shift

    I_edges[0] = x_min
    I_edges[-1] = x_max

    return I_edges

labels = ["muy bajo", "bajo", "medio", "alto", "muy alto"]

bins_wf = weber_fechner_bins(df["SUMA"], n_clases=5)
bins_val_wf = weber_fechner_bins(df["SUMA_val"], n_clases=5)

df["indice_wf"] = pd.cut(
    df["SUMA"],
    bins=bins_wf,
    labels=labels,
    include_lowest=True
)

df["indice_val_wf"] = pd.cut(
    df["SUMA_val"],
    bins=bins_val_wf,
    labels=labels,
    include_lowest=True
)

df
```
